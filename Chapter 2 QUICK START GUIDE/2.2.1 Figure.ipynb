 The whole figure. The Figure keeps track of all the child Axes, a group of 'special' Artists (titles, figure
 legends, colorbars, etc.), and even nested subfigures.

 Typically, you'll create a new Figure through one of the following functions:

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
fig = plt.figure()  # an empty figure with no Axes

```

```python
fig, ax = plt.subplots() # a figure with a single Axes
```

```python
fig, axs = plt.subplots(2, 2) # a figure with a 2x2 grid of Axes
```

```python
# a figure with one Axes on the left, and two on the right:
fig, axs = plt.subplot_mosaic([['left', 'right_top'],['left', 'right_bottom']])
```
