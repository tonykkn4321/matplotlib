```python
import matplotlib.pyplot as plt
import numpy as np
```

Matplotlib graphs your data on Figures (e.g., windows, Jupyter widgets, etc.), each of which can contain one or more Axes,an area where points can be specified in terms of x-ycoordinates (or theta-r in a polar plot,  x-y-z in a 3D plot, etc.).

The simplest way of creating a Figure with an Axes is using pyplot.subplots.
 Wecan then use Axes.plot to draw some data on the Axes, and show to display the figure:

```python
fig, ax = plt.subplots()             # Create a figure containing a single Axes.
ax.plot([1, 2, 3, 4], [1, 4, 2, 3])  # Plot some data on the Axes.
plt.show()
```
